```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.amazon.in/Airdopes-441-Technology-Immersive-Resistance/product-reviews/B084DS51NC/ref=cm_cr_getr_d_paging_btm_prev_1?pageNumber=1'
data=requests.get(url)
```

```python
soup = BeautifulSoup(data.content,'html.parser')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
review_title = soup.select('a.review-title')
review_title
```

```python
len(review_title)
```

```python
ratings = soup.select('i.review-rating span.a-icon-alt')[2:]
ratings
```

```python
len(ratings)
```

```python
reviews = soup.select('span.review-text-content span')
reviews
```

```python
len(reviews)
```

```python
ratings_l = []
review_title_l = []
reviews_l = []
for i in range(10):
  ratings_l.append(ratings[i].get_text())
  review_title_l.append(review_title[i].get_text())
  reviews_l.append(reviews[i].get_text().strip("\n "))
```

```python
ratings_l
```

```python
review_title_l
```

```python
reviews_l
```

```python
import pandas as pd

df=pd.DataFrame({'Review_Title':review_title_l,'Reviews':reviews_l,'Rating':ratings_l})
```

```python
df
```

```python
df['Review_Title']
```

```python
import re
```

```python
def remove(text):
  text=re.sub("\n",'',text)
  return text
```

```python
df['Review_Title']=df['Review_Title'].apply(remove)
```

```python
def remove_rating(text):
  text=re.sub("out of 5 stars",'',text)
  return text
```

```python
df['Rating']=df['Rating'].apply(remove_rating)
```

```python
df
```

```python
ratings_l = []
review_title_l = []
reviews_l = []
for page in range(1,450):
  url = "https://www.amazon.in/Airdopes-441-Technology-Immersive-Resistance/product-reviews/B084DS51NC/ref=cm_cr_getr_d_paging_btm_prev_1?pageNumber="+str(page)
  data = requests.get(url)
  if str(data) == "<Response [200]>":
    soup = BeautifulSoup(data.content,'html.parser')
    review_title = soup.select('a.review-title')
    ratings = soup.select('i.review-rating span.a-icon-alt')[2:]
    reviews = soup.select('span.review-text-content span')
    for i in range(10):
      ratings_l.append(ratings[i].get_text())
      review_title_l.append(review_title[i].get_text())
      reviews_l.append(reviews[i].get_text().strip("\n "))
```

```python
df1=pd.DataFrame({'Review_Title':review_title_l,'Reviews':reviews_l,'Rating':ratings_l})
```

```python
df1
```

```python
df1.shape
```

```python
df1['Review_Title']=df1['Review_Title'].apply(remove)
```

```python
df1['Rating']=df1['Rating'].apply(remove_rating)
```
